```python
class TriangleChecker:
    
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
        
    def is_triangle(self):
        if any([((type(i) != int) and type(i) != float) or (i <= 0) for i in [self.a, self.b, self.c]]):
            return 1
        elif self.a + self.b <= self.c or self.a + self.c <= self.b or self.b + self.c <= self.a:
            return 2
        else:
            return 3
            

a, b, c = map(int, input().split())
tr = TriangleChecker(a, b, c)
print(tr.is_triangle())
```

```python
from string import ascii_lowercase, digits

class CardCheck:
    NUMBERS = digits
    NFILEDS = 4
    LENFIELD = 4
    CHARS_FOR_NAME = ascii_lowercase.upper() + digits
    NWORDS = 2
    
    @classmethod
    def check_card_number(cls, number):
        n_fields_check = len(number.split('-')) == cls.NFILEDS
        len_fields_check = all([len(i) == cls.LENFIELD for i in number.split('-')])
        digits_check = all([all([i in cls.NUMBERS for i in n]) for n in number.split('-')])
        return n_fields_check and len_fields_check and digits_check
    
    @classmethod
    def check_name(cls, name):
        n_words_check = len(name.split()) == 2
        chars_chek = all([all([i in cls.CHARS_FOR_NAME for i in w]) for w in name.split()])
        return n_words_check and chars_chek
```

```python
class Data:
    
    def __init__(self, data, ip):
        self.data = str(data)
        self.ip = int(ip)
        print(f"Data: {self.data}")
        print(f"Reciever ip: {self.ip}")
    
    
class Server:
    IP_LIST = []
    #ROUTER = None
    
    @classmethod
    def pick_ip(cls):
        if len(cls.IP_LIST) == 0:
            cls.IP_LIST.append(1)
            return 1
        elif cls.IP_LIST == list(range(1,len(cls.IP_LIST)+1)):
            ip_buff = len(cls.IP_LIST)+1
            cls.IP_LIST.append(ip_buff)
            return ip_buff
        else:
            ip_buff = 0
            ip_range = list(range(1,cls.IP_LIST[-1]+1))
            for ip, ip_gap in zip(cls.IP_LIST, ip_range):
                if ip != ip_gap:
                    ip_buff = ip_gap
                    break
            cls.IP_LIST.append(ip_buff)
            cls.IP_LIST = sorted(cls.IP_LIST)
            return ip_buff
            
    
    def __init__(self):
        self.ip = self.pick_ip()
        self.buffer = []
        self.router = None
        self.data_sended = False
        print(f"Server created, ip = {self.ip}")
        
    def get_ip(self):
        return self.ip
    
    def send_data(self, data):
        if not self.router is None:
            self.router.buffer.append(data)
            print(f"Data: |{data.data}| sent from ip = {self.ip} to ip = {data.ip}")
        else:
            print('Server has no connection with router')
            
    def get_data(self):
        data_buff = [i for i in self.buffer]
        self.buffer.clear()
        return data_buff
        
    
    
class Router:
    
    def __init__(self):
        self.buffer = []
        self.linked_servers = []
        
    def link(self, server):
        if not server in self.linked_servers:
            self.linked_servers.append(server)
            server.router = self
            print(f"Router linked server with ip = {server.ip}")
            
    def unlink(self, server):
        if server in self.linked_servers:
            self.linked_servers.remove(server)
            server.router = None
            print(f"Router unlinked server with ip = {server.ip}")
            
    def send_data(self):
        for server in self.linked_servers:
            for data in self.buffer:
                if server.ip == data.ip:
                    server.buffer.append(data)
                    print(f"Router sent |{data.data}| to  ip = {server.ip}")
                else:
                    continue
        self.buffer.clear()                    
```

```python
# Function decorator
def func_decorator(func):
    def wrapper(*args, **kwargs):
        print('-------- Befor calling the func ----------')
        res = func(*args, **kwargs)
        print('-------- After calling the func ----------')
        return res
    return wrapper

def some_func(title):
    print(f'Calling some_func {title}')
    return f'<tag>{title}</tag>'
    
some_func = func_decorator(some_func)
f = some_func('Title')
print(f)
```

```python
import time

# Function decorator
def test_time(func):
    def wrapper(*args, **kwargs):
        st = time.time()
        res = func(*args, **kwargs)
        et = time.time()
        dt = et - st
        print(f'time of calculations = {dt} sec')
        return res
    return wrapper

@test_time
def get_nod(a, b):
    while a != b:
        if a > b:
            a -= b
        else:
            b -= a
        return a

@test_time
def get_fast_nod(a, b):
    if a < b:
        a, b = b, a
    while b:
        a, b = b, a % b
    return a
    
#get_nod = test_time(get_nod)
#get_fast_nod = test_time(get_fast_nod)

res = get_nod(54, 66666)
res2 = get_fast_nod(54, 66666)
print(res, res2)
```

```python
import math
from functools import wraps

def df_decorator(dx=0.01):
    def func_decorator(func):
        @wraps(func)
        def wrapper(x, *args, **kwargs):
            # we want to choose this parameter
            #dx = 0.00001
            res = (func(x+dx, *args, **kwargs) - func(x, *args, **kwargs)) / dx
            return res
        # wrapper.__name__ = func.__name__
        # wrapper.__doc__ = func.__doc__
        return wrapper
    return func_decorator

@df_decorator(dx=0.001)
def sin_df(x):
    """Function calculates derivative of sin"""
    return math.sin(x)

df = sin_df(math.pi/3)
print(df)
print(sin_df.__name__)
print(sin_df.__doc__)
```

```python
class StackObj:
    def __init__(self, data):
        self.__data = data
        self.__next = None
        print(f"StackObj with data = {self.data} created")
        
    @property
    def next(self):
        return self.__next
    
    @next.setter
    def next(self, next_obj):
        if type(next_obj) == StackObj or next_obj is None:
            self.__next = next_obj
            
    @next.deleter
    def next(self):
        del self.__next
        self.__next = None
            
    @property
    def data(self):
        return self.__data
    
    @data.setter
    def data(self, data):
        if type(data) == str:
            self.__data = data
            
    
class Stack:
    def __init__(self):
        self.top = None
        
    def get_last(self):
        last_obj = self.top
        counter = 0
        while last_obj.next != None:
            counter += 1
            print(f"{counter}. data = {last_obj.data}")
            last_obj = last_obj.next
        counter += 1
        print(f"{counter}. data = {last_obj.data}")
        print(f"Found last object in Stack with data = {last_obj.data}")
        return last_obj
    
    def get_before_last(self):
        last_obj = self.top
        if last_obj.next != None:
            while last_obj.next.next != None:
                last_obj = last_obj.next
        else:
            pass
        print(f"Found last but one object in Stack with data = {last_obj.data}")
        return last_obj
        
    def push(self, obj):
        if self.top == None:
            self.top = obj
            print(f"First object with data = {self.top.data} added to Stack")
        else:
            last_obj = self.get_last()
            last_obj.next = obj
            print(f"Here in push last object was {last_obj.data}")
            print(f"Object with data = {obj.data} added to Stack")
        print(f"Now last object is {self.get_last().data}")
        print(f"TOP object is {self.top.data}")
        
    def pop(self):
        new_last_obj = self.get_before_last()
        if new_last_obj is self.top:
            print(f"NEW last object is TOP {self.top.data} ???")
            buff_obj = new_last_obj
            del new_last_obj
            self.top = None
            return buff_obj
        else:    
            print(f"Delete last object from Stack with data = {new_last_obj.next.data}")
            buff_obj = new_last_obj.next
            del new_last_obj.next
            return buff_obj
        #else:
        #    print(f"Last object is TOP with data = {new_last_obj.data}")
        #    buff_obj = new_last_obj
        #    del new_last_obj
        #    return buff_obj
        
    def get_data(self):
        list_str = []
        if not self.top == None:
            list_str.append(self.top.data)
            if not self.top.next is None:
                last_obj = self.top.next
                while last_obj != None:
                    list_str.append(last_obj.data)
                    last_obj = last_obj.next
        return list_str
    
st = Stack()
st.push(StackObj("obj1"))
st.push(StackObj("obj2"))
st.push(StackObj("obj3"))
st.pop()
res = st.get_data()    # ['obj1', 'obj2']
print(res)
```

```python
class RadiusVector2D:
    MIN_COORD = -100
    MAX_COORD = 1024
    
    def __init__(self, x=0, y=0):
        if type(x) in (int, float) and self.MIN_COORD <= x <= self.MAX_COORD:
            self.__x = x
        else:
            self.__x = 0
        if type(y) in (int, float) and self.MIN_COORD <= y <= self.MAX_COORD:
            self.__y = y
        else:
            self.__y = 0
        
    @property
    def x(self):
        return self.__x
    
    @x.setter
    def x(self, x_new):
        if type(x_new) in (int, float) and self.MIN_COORD <= x_new <= self.MAX_COORD:
            self.__x = x_new
        else:
            pass
        
    @property
    def y(self):
        return self.__y
    
    @y.setter
    def y(self, y_new):
        if type(y_new) in (int, float) and self.MIN_COORD <= y_new <= self.MAX_COORD:
            self.__y = y_new
        else:
            pass
        
    @staticmethod
    def norm2(vector):
        return vector.x**2 + vector.y**2
```

```python
import math

class LineTo:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        self.len = 0
        
    def set_end_coords(self, x_st=0,y_st=0):
        self.x0 = x_st
        self.y0 = y_st
        self.len = math.sqrt((self.x - self.x0)**2 + (self.y - self.y0)**2)
        
    def get_end_coords(self):
        return self.x, self.y
    
    def get_len(self):
        return self.len
    

class PathLines:
    def __init__(self, *args):
        self.lines_list = []
        for i in range(0,len(args)):
            if i == 0:
                args[i].set_end_coords()
            else:
                args[i].set_end_coords(*(args[i-1].get_end_coords()))
            self.lines_list.append(args[i])
            
    def get_path(self):
        return self.lines_list
    
    def get_length(self):
        length = 0
        for line in self.lines_list:
            length += line.get_len()
        return length
    
    def add_line(self, line):
        if len(self.lines_list) == 0:
            line.set_end_coords()
        else:
            line.set_end_coords(*(self.lines_list[-1].get_end_coords()))
        self.lines_list.append(line)

        
p = PathLines(LineTo(10, 20), LineTo(10, 30))
p.add_line(LineTo(20, -10))
dist = p.get_length()
print(dist)
```

```python
# Data descriptor
class FloatValue:

    @classmethod
    def verifycoord(cls, coord):
        if type(coord) != float:
            raise TypeError("Присваивать можно только вещественный тип данных.")

    def __set_name__(self, owner, name):
        self.name = '_' + name

    def __get__(self, instance, owner):
        return getattr(instance, self.name)

    def __set__(self, instance, value):
        self.verifycoord(value)
        setattr(instance, self.name, value)
        
class Cell:
    
    value = FloatValue()
    
    def __init__(self, value=0.0):
        self.value = value
        
        
class TableSheet:
    def __init__(self, N, M):
        self.N = N
        self.M = M
        self.cells = [[Cell() for i in range(1,M+1)] for j in range(1,N+1)]
        
table = TableSheet(5, 3)
table.cells = [[Cell(float(i)+float(j-1)*table.M) for i in range(1,table.M+1)] for j in range(1,table.N+1)]
for i in table.cells:
    row = [j.value for j in i]
    print(row)
```

```python
class ValidateString:
    def __init__(self, min_length, max_length):
        self.min_length = min_length
        self.max_length = max_length
        
    def validate(self, string):
        return type(string) == str and self.min_length <= len(string) <= self.max_length


class StringValue:
    def __init__(self, validator):
        self.validator = validator
    
    def __set_name__(self, owner, name, *kwargs):
        self.name = '_' + name
        
    def __get__(self, instance, owner):
        return getattr(instance, self.name)
        
    def __set__(self, instance, str_value):
        if self.validator.validate(str_value):
            setattr(instance, self.name, str_value)
            
    
class RegisterForm:
    
    login = StringValue(validator=ValidateString(min_length=3, max_length=100))
    password = StringValue(validator=ValidateString(min_length=3, max_length=100))
    email = StringValue(validator=ValidateString(min_length=3, max_length=100))
    
    def __init__(self, login_str, password_str, email_str):
        self.login = login_str
        self.password = password_str
        self.email = email_str
        
    def get_fields(self):
        return [self.login, self.password, self.email]
    
    def show(self):
        print(f'<form>\nЛогин: {self.login}\nПароль: {self.password}\nEmail: {self.email}\n</form>')
```

```python
class Thing:
    def __init__(self, name, weight):
        self.weight = weight
        self.name = name


class Bag:
    def __init__(self, max_weight):
        self.max_weight = max_weight
        self.__things = []
            
    @property
    def things(self):
        return self.__things
            
    def get_total_weight(self):
        return sum([i.weight for i in self.things])
            
    def add_thing(self, thing):
        if (self.get_total_weight() + thing.weight) <= self.max_weight:
            self.things.append(thing)
        else:
            pass
        
    def remove_thing(self, indx):
        del self.things[indx]
```

```python
class Telecast:
    def __init__(self, prog_id, name, duration):
        self.__id = prog_id
        self.__name = name
        self.__duration = duration
        
    @property
    def uid(self):
        return self.__id
    
    @uid.setter
    def uid(self, prog_id):
        if type(prog_id) == int and prog_id >= 1:
            self.__id = prog_id
        else:
            pass
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, name):
        if type(name) == str:
            self.__name = name
        else:
            pass
    
    @property
    def duration(self):
        return self.__duration
    
    @duration.setter
    def duration(self, duration):
        if type(duration) == int and duration > 0:
            self.__duration = duration
        else:
            pass

        
class TVProgram:
    def __init__(self, channel_name):
        self.name = channel_name
        self.items = []
        
    def add_telecast(self, tl):
        self.items.append(tl)
        
    def remove_telecast(self, indx):
        for tl in self.items:
            if tl.uid == indx:
                self.items.remove(tl)
                
pr = TVProgram("Первый канал")
pr.add_telecast(Telecast(1, "Доброе утро", 10000))
pr.add_telecast(Telecast(2, "Новости", 2000))
pr.add_telecast(Telecast(3, "Интервью с Балакиревым", 20))
pr.remove_telecast(2)
for t in pr.items:
    print(f"{t.name}: {t.duration}")
```

```python
# неверное решение
class Value2D:

    @classmethod
    def verifycoord(cls, coord):
        if type(coord) not in [int, float]:
            raise TypeError("Неверный тип присваиваемых данных.")
            
    def verifyradius(cls, radius):
        if type(radius) not in [int, float] or radius <= 0:
            raise TypeError("Неверный тип присваиваемых данных.")

    def __set_name__(self, owner, name):
        self.name = '__' + name

    def __get__(self, instance, owner):
        # return instance.__dict__[self.name]
        return getattr(instance, self.name)

    def __set__(self, instance, value):
        self.verifycoord(value)
        if self.name == 'radius':
            self.verifyradius(value)
        # instance.__dict__[self.name] = value
        setattr(instance, self.name, value)


class Circle:
    x = Value2D()
    y = Value2D()
    radius = Value2D()

    def __init__(self, x, y, radius):
        self.x = x
        self.y = y
        self.radius = radius
        
    def __getattr__(self, item):
        return False
        
        
circle = Circle(10.5, 7, 22)
circle.radius = -10 # прежнее значение не должно меняться, т.к. отрицательный радиус недопустим
x, y = circle.x, circle.y
res = circle.name # False, т.к. атрибут name не существует        
```

```python
class Dimensions:
    MIN_DIMENSION = 10
    MAX_DIMENSION = 1000
    
    def __init__(self, a, b, c):
        self.__a = a
        self.__b = b
        self.__c = c
        
    def __setattr__(self, key, value):
        print(key, value)
        print(value < self.MIN_DIMENSION or value > self.MAX_DIMENSION)
        if not type(value) in [int, float] or (value < self.MIN_DIMENSION or value > self.MAX_DIMENSION):
            print(f'trouble with {key} = {value}')
            pass
        elif key in ['MIN_DIMENSION', 'MAX_DIMENSION']:
            raise AttributeError("Менять атрибуты MIN_DIMENSION и MAX_DIMENSION запрещено.")
        else:
            object.__setattr__(self, key, value)
        
    @property
    def a(self):
        return self.__a
    
    @a.setter
    def a(self, a):
        self.__a = a
        
    @property
    def b(self):
        return self.__b
    
    @b.setter
    def b(self, b):
        self.__b = b
        
    @property
    def c(self):
        return self.__c
    
    @c.setter
    def c(self, c):
        self.__c = c
        
        
d = Dimensions(10.5, 20.1, 30)
d.a = 8
d.b = -20
a, b, c = d.a, d.b, d.c  # a=10.5, b=15, c=30
print(d.__dict__)
```

```python
import time

class Mechanical:
    def __init__(self, date):
        self.date = date
        
    def __setattr__(self, key, value):
        if key == 'date' and key in self.__dict__.keys():
            pass
        else:
            object.__setattr__(self, key, value)
            
class Aragon:
    def __init__(self, date):
        self.date = date
        
    def __setattr__(self, key, value):
        if key == 'date' and key in self.__dict__.keys():
            pass
        else:
            object.__setattr__(self, key, value)
            
            
class Calcium:
    def __init__(self, date):
        self.date = date
        
    def __setattr__(self, key, value):
        if key == 'date' and key in self.__dict__.keys():
            pass
        else:
            object.__setattr__(self, key, value)


class GeyserClassic:
    MAX_DATE_FILTER = 100
    
    def __init__(self):
        self.filters = {1: None, 2: None, 3: None}
        
    def add_filter(self, slot_num, filter_type):
        if self.filters[slot_num] == None:
            if slot_num == 1 and isinstance(filter_type, Mechanical):
                self.filters[slot_num] = filter_type
            elif slot_num == 2 and isinstance(filter_type, Aragon):
                self.filters[slot_num] = filter_type
            elif slot_num == 3 and isinstance(filter_type, Calcium):
                self.filters[slot_num] = filter_type
    
    def remove_filter(self, slot_num):
        self.filters[slot_num] = None
        
    def get_filters(self):
        return tuple(self.filters.values())
    
    def time_ok(self):
        if None in self.get_filters():
            return False
        else:
            return all([0 <(time.time() - f.date) <= self.MAX_DATE_FILTER for f in self.filters.values()])
    
    def water_on(self):
        return not None in self.get_filters() and self.time_ok()
```

```python
my_water = GeyserClassic()
my_water.add_filter(1, Mechanical(time.time()))
my_water.add_filter(2, Aragon(time.time()))
w = my_water.water_on() # False
my_water.add_filter(3, Calcium(time.time()))
w = my_water.water_on() # True
f1, f2, f3 = my_water.get_filters()  # f1, f2, f3 - ссылки на соответствующие объекты классов фильтров
my_water.add_filter(3, Calcium(time.time())) # повторное добавление в занятый слот невозможно
my_water.add_filter(2, Calcium(time.time())) # добавление в "чужой" слот также невозможно
```
